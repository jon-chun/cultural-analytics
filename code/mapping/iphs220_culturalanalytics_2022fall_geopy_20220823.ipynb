## Explore geopy.geolocator functionality

```python
geolocator = Nominatim(user_agent="Kenyon IPHS220 Fall2022 GeoPy Demo", timeout=2)
```

```python
location = geolocator.geocode("Kenyon College")
```

```python
location
```

```python
type(location)
```

```python
print(location.address)
```

```python
dir(location)
```

```python
print(location.latitude, location.longitude)
```

```python
print(f"Importance: {location.raw['importance']}")
```

```python
type(location.raw)
```

```python
print(f"Class: {location.raw['class']} \nType: {location.raw['type']}")
```

```python
possible_locations = geolocator.geocode("Washington", exactly_one=False)

for location in possible_locations:
    print(location.address)
    print(location.latitude, location.longitude)
    print(f"Importance: {location.raw['importance']}")
```

```python
location = geolocator.geocode("Chase Ave, Gambier, OH")

print(location.address)
print(location.latitude, location.longitude)
print(f"Importance: {location.raw['importance']}")
```

```python
po_coords = geolocator.geocode("Post Office, Gambier, OH")

print(po_coords.address)
print(po_coords.latitude, po_coords.longitude)
print(f"Importance: {po_coords.raw['importance']}")
```
